```python
#FOR K = 2 (ORIGINAL FORMULA)
import numpy as np
import matplotlib.pyplot as plt

x = 25

x_values = list(range(1, x+1))
p_values = []
p_x = 0

for i in range(1, x+1):
    p_x = 1 - 0.5*p_x
    p_values.append(p_x)

plt.figure(figsize=(10, 6))
plt.plot(x_values, p_values, marker='o', linestyle='', color='dodgerblue')
plt.axhline(y=2/3, color='red', linestyle='--', label=r'$\frac{2}{3}$ (limit)')

plt.title('Convergence of $P(x)$ to $\dfrac{2}{3}$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('$P(x)$', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#FOR K = 3 (ORIGINAL FORMULA)
import numpy as np
import matplotlib.pyplot as plt

x = 25

x_values = list(range(1,x+1))
p_values = [1,1/3]

for i in range(x-2):
    p_x = 1 - (2/3)*p_values[i+1] - (1/3)*p_values[i]
    p_values.append(p_x)

plt.figure(figsize=(10, 6))
plt.plot(x_values, p_values, marker='o', linestyle='', color='dodgerblue')
plt.axhline(y=1/2, color='red', linestyle='--', label=r'$\frac{1}{2}$ (limit)')

plt.title('Convergence of $P(x,3)$ to $\dfrac{1}{2}$', fontsize=14)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$P(x,3)$', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#FOR K = 4 (ORIGINAL FORMULA)
import numpy as np
import matplotlib.pyplot as plt

x = 25

x_values = list(range(1,x+1))
p_values = [1,1/4,5/16]

for i in range(x-3):
    p_x = 1 - (3/4)*p_values[i+2] - (1/2)*p_values[i+1] - (1/4)*p_values[i]
    p_values.append(p_x)
    
lim = p_values[x-1]   

plt.figure(figsize=(10, 6))
plt.plot(x_values, p_values, marker = 'o', linestyle='', color='dodgerblue')
plt.axhline(y=lim, color='red', linestyle='--', label=r'$\frac{2}{5}$ (limit)')

plt.title('Convergence of $P(x,4)$ to $\dfrac{2}{5}$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('$P(x,4)$', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
