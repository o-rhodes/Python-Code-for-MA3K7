```python
def B(x, y, length=1000):
    bracelet = [x, y]
    
    # Generate the bracelet sequence until the pair (x, y) repeats
    for i in range(2, length):
        next_num = (bracelet[i-1] + bracelet[i-2]) % 10
        bracelet.append(next_num)
        
        # Check if the pair (bracelet[i-1], bracelet[i]) repeats the initial pair (x, y)
        if bracelet[i-1] == x and bracelet[i] == y:
            break
    
    return bracelet

# Loop through all combinations of x and y where both vary from 0 to 9
for l in range(10):
    for k in range(10):
        x = l  # First number varies from 0 to 9
        y = k  # Second number varies from 0 to 9
    
        # Generate the bracelet sequence
        bracelet = B(x, y)
    
        # Print the bracelet sequence and the length excluding the first two numbers
        print(f"Generated bracelet sequence for x={x}, y={y}:")
        print(bracelet)
        print(f"Length (before it repeats): {len(bracelet) - 2}")
        print()  # Just to separate the output of each iteration
```
